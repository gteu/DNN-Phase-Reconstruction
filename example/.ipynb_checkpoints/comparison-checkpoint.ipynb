```python
import librosa
import lws
import numpy as np
import IPython.display
import matplotlib.pyplot as plt
import librosa.display

SR = 16000
N_FFT = 512
HOP_LEN = 80
WIN_LEN = 400

def show_audio(y):
    filename = "tmp.wav"
    librosa.output.write_wav(filename, y, sr=SR)
    IPython.display.display(IPython.display.Audio(filename))
```

```python
test_filename = "test.wav"
y, _ = librosa.load(test_filename, sr=SR, mono=True)
show_audio(y)
```

```python
lws_processor=lws.lws(WIN_LEN, HOP_LEN, mode="speech")
X = lws_processor.stft(y) 

X0 = np.abs(X) # Magnitude spectrogram
X1 = lws_processor.run_lws(X0) # reconstruction from magnitude

# y_r = lws_processor.istft(np.transpose(X1))
y_r = librosa.istft(np.transpose(X1), hop_length=HOP_LEN, win_length=WIN_LEN)
show_audio(y_r)
```
